```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
# read csv dataset
dataset = {}
for i in range(17,24):
    dataset["Y20"+str(i)] = pd.read_csv('FIFA'+str(i)+'_official_data.csv')
```

```python
# tidy up dataset
dataset["Y2023"].info()
```

```python
dataset["Y2023"].head(10)
```

```python
print(dataset['Y2023'].columns.values)
```

```python
columns_to_drop = ['ID', 'Photo', 'Nationality', 'Flag', 'Club Logo', 'International Reputation', 'Real Face', 'Position', 'Joined', 'Loaned From', 'Contract Valid Until', 'Best Overall Rating']
for i in range(17,24):
    dataset['Y20'+str(i)] = dataset['Y20'+str(i)].drop(columns=columns_to_drop,axis=1)
dataset["Y2023"].info()
```

```python
# ADD COLUMNS
weight = [int(i[0:2]) for i in df23["Weight"]]
height = [int(i[0:3]) for i in df23["Height"]]
df23['BMI'] = [round(w/(h/100)**2,2) for w, h in zip(weight, height)] 
```

```python
df23.drop(['Weak Foot','Skill Moves','Body Type','Kit Number'], axis=1, inplace=True)
```

```python
merge = []
for i in range(17,24):
    dataset['Y20'+str(i)]["Year"] = "20"+str(i)
    merge.append(dataset["Y20"+str(i)])
dataset = pd.concat(merge)
dataset.groupby(['Name']).mean()
dataset
```

```python
labels =['Age','Overall','Potential','Value','Wage']
dataset = dataset[labels]

numb = len(labels)
dataset
values = list(dataset.loc[1])
angles = np.linspace(0,2*np.pi, numb, endpoint=False).tolist()

# PLot is a circle, we need the loop and append the start to the end
values += values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10,10),subplot_kw=dict(polar=True))

# Fix axis to go in the right order and start at 12 o'clock
ax.set_theta_offset(np.pi/2)
ax.set_theta_direction(-1)

#draw axis line for each angle and label
# ax.set_theta_grids(np.degrees(angles), labels)

# draw outline for data
ax.plot(angles, values, color='red', linewidth=1)

ax.fill(angles, values,color='red',alpha=0.25)
ax.set_title('Radar Plot')
```
